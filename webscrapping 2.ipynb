```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.mahindrafirstchoice.com/used-cars/mumbai"
headers = {
    "useragent": "Mozilla/5.0+(Windows+NT+10.0;+Win64;+x64)+AppleWebKit/537.36+(KHTML,+like+Gecko)+Chrome/92.0.4515.107+Safari/537.36"
}
```

```python
request=requests.get(url,headers=headers)
soup=BeautifulSoup((request).text,"html.parser")
```

```python
soup
```

```python
car_list=soup.select('#buyer_stock_results')[0]
rows=soup.select('.general-car')
result=[]

for row in rows:
    info=soup.select('.buyer_make')[0].text.split()
    year=info[0]
    company=info[1]
    model=soup.select('.buyer_variant')[0].text.strip()
    price=soup.select('.car_price')[0].text.strip()
    location=soup.select('.buyer_locations')[0].text.strip()
    specs=soup.select('.spec-info-list')[0].select("li")
    km=specs[0].select('.overview_name')[0].text.replace(",","")
    type_car=specs[1].select('.overview_name')[0].text.strip()
    body_type=specs[2].select('.overview_name')[0].text
    owner=specs[3].select('.overview_name')[0].text
    
    data = {
        "year": year,
        "company": company,
        "model":model,
        "price":price,
        "buyer location":location,
        "kilometer":km,
        "type of car":type_car,
        "body type":body_type,
        "owner":owner
    }
    result.append(data)
    
result
```

```python
import csv
```

```python
fields=['year','company','model','price','buyer location','kilometer','type of car','body type','owner']

filename = "car_mumbai.csv"
  
# writing to csv file
with open(filename, 'w') as csvfile:
    # creating a csv dict writer object
    writer = csv.DictWriter(csvfile, fieldnames = fields)
      
    # writing headers (field names)
    writer.writeheader()
      
    # writing data rows
    writer.writerows(result)
```
